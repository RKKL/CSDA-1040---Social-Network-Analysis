## 2.1 Initial Dataset Import and Configuration

```python

fbdf = pd.read_excel(u'Facebook_Data_clean.xlsx',index_col=0)
fbdf
```

```python
print(len(fbdf.columns))
```

```python
# The first 45 columns, have the utf-8 character 'xa0' (double space) appended to the column name,
# and need to remove it

fbdf.columns = fbdf.columns.str.replace(u'xa0', u'', regex=True)
```

```python
# In case there are columns with no names, remove them
fbdf = fbdf.dropna(axis=1,how='all')
```

```python
fbdf.columns
```

```python
# How many columns?
print(len(fbdf.columns))
```

### Create the graph using a function from networkx (nx). The graph can be directed or undirected

```python
# Create undirected grapth
# FBG = nx.from_pandas_adjacency(fbdf)
```

```python
# For Directed grapth must explicitly mention this in the creation
FBG = nx.from_pandas_adjacency(fbdf, create_using=nx.DiGraph)
```

```python
FBG.name = "Facebook Graph (FBG) created from the pandas FB adjacency matrix"
print(nx.info(FBG))
```
